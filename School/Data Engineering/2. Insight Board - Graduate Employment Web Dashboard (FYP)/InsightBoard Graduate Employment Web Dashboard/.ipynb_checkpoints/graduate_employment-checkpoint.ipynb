```python
# First, we convert all the numerical data columns to numeric type, setting errors='coerce' to handle non-numeric values
columns_to_convert = ['employment_rate_overall', 'employment_rate_ft_perm', 
                      'basic_monthly_mean', 'basic_monthly_median', 
                      'gross_monthly_mean', 'gross_monthly_median', 
                      'gross_mthly_25_percentile', 'gross_mthly_75_percentile']

for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Now let's plot the distribution of these columns using histograms
fig, axes = plt.subplots(4, 2, figsize=(15, 20))  # Adjusting the subplot grid for better visualization

axes = axes.ravel()  # Flatten the array for easy iteration

for i, column in enumerate(columns_to_convert):
    sns.histplot(df[column].dropna(), ax=axes[i], kde=True, bins=30)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

-Employment Rates (employment_rate_overall, employment_rate_ft_perm):
- These might show a skewed distribution towards higher values, which is common for employment rates, suggesting that most graduates find employment. A concentration towards 100% could indicate high employment success among the surveyed graduates.
    
-Basic Monthly Mean and Median (basic_monthly_mean, basic_monthly_median):
- Salary distributions can often be right-skewed, meaning there's a tail towards the higher salaries, with most graduates earning within a lower to mid-salary range. The mean might be higher than the median if there are outliers with very high salaries.

-Gross Monthly Mean and Median (gross_monthly_mean, gross_monthly_median):
- Similar to basic monthly salaries, gross monthly salaries are likely right-skewed. The presence of higher earning outliers can pull the mean up, making it larger than the median. The bulk of the data might cluster around a central range that represents the common earning bracket.

-Gross Monthly 25th and 75th Percentiles (gross_mthly_25_percentile, gross_mthly_75_percentile):
- These columns provide insights into the distribution's spread. The 25th percentile represents the lower quartile, below which 25% of the observations fall, and the 75th percentile represents the upper quartile, above which 25% of the observations fall. The distance between these values (the interquartile range) can give an idea of the data's spread. A larger range indicates more variability in graduate salaries.

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your loaded DataFrame

# Convert "na" strings to NaN across the entire DataFrame
df = df.replace('na', np.nan, inplace=False)

# Convert columns to numeric where necessary. This is essential because
# the median operation requires the columns to be in a numeric format.
numeric_columns = ['employment_rate_overall', 'employment_rate_ft_perm', 
                   'basic_monthly_mean', 'basic_monthly_median', 
                   'gross_monthly_mean', 'gross_monthly_median', 
                   'gross_mthly_25_percentile', 'gross_mthly_75_percentile']

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Fill "NaN" for specified columns with their median values
for column in numeric_columns:
    df[column].fillna(df[column].median(), inplace=True)

```
