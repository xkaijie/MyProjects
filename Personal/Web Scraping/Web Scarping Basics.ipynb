## Web Scarping
### Url Lib

```python
#urllib
import urllib.request
```

```python
#headers are not included here
urr = urllib.request.urlopen('https://www.port.ac.uk/')
content = urr.read()
urr.close()
html = content.decode()
print(html)
```

### Request.get

```python
import requests
rer = requests.get('https://www.port.ac.uk/')
print(rer.status_code)
html = rer.text
print(html) 
```

`urllib` and `requests.get` are both Python libraries used for making HTTP requests, but they have some differences in terms of functionality and ease of use:

1. Ease of use: `requests.get` from the `requests` library provides a higher-level and more user-friendly interface compared to `urllib`. It offers a simplified API and abstracts away some of the complexities of making HTTP requests, making it easier to work with.

2. Convenience features: `requests` provides additional features and convenience methods that are not available in `urllib`. For example, it offers built-in support for handling cookies, handling redirects, automatic content decoding, session management, and more. These features make it more convenient to work with web APIs and perform common tasks related to HTTP requests.

3. Code readability: The `requests` library is known for its clear and concise syntax, which enhances code readability. The code written with `requests.get` tends to be more concise and easier to understand compared to the equivalent code written with `urllib`.

4. Third-party libraries: `requests` has gained popularity in the Python community and has become the de facto standard for making HTTP requests. As a result, many third-party libraries and frameworks integrate and provide support specifically for `requests`, making it more widely compatible and extensible.

5. Compatibility: `urllib` is a built-in library in Python, available in both Python 2 and Python 3 by default. On the other hand, `requests` is a third-party library that needs to be installed separately, although it is widely supported and compatible with both Python 2 and Python 3.

6. Advanced functionality: While `requests` provides a high-level API that covers most common use cases, `urllib` offers a broader set of functionalities for more advanced use cases. `urllib` provides modules like `urllib.request`, `urllib.parse`, `urllib.error`, and `urllib.robotparser`, which allow fine-grained control over the HTTP requests, handling URLs, handling errors, parsing URLs, and parsing robots.txt files, respectively.

In summary, `requests.get` is often preferred over `urllib` for making HTTP requests in Python due to its simplicity, convenience, and widespread adoption. It offers a more user-friendly interface and additional features that make working with HTTP requests easier and more efficient. However, in certain cases where more advanced functionality or fine-grained control is required, `urllib` might be a suitable choice.

## Beautiful Soup - extract with re(regular expression)

```python
pip install lxml
```

```python
html_doc = """
    <html><head><title>The Dormouse's story</title></head>
    <body>
    <p class="title"><b>The Dormouse's story</b></p>
    <p class="story">Once upon a time there were three little sisters; and their names were
    <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
    <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
    <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
    and they lived at the bottom of a well.</p>
    <p class="story">...</p>
"""

# you will need to install the lxml parser with $ pip install lxml
from bs4 import BeautifulSoup as bs
import re

soup = bs(html_doc, 'lxml')
print(soup.prettify())

for link in soup.find_all('a'):
    print(link.get('href'))

print(soup.get_text())

for tag in soup.find_all(re.compile("^b")):
    print(tag.name)
```
