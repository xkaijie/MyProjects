Operators are special symbols in Python that carry out arithmetic or logical computation. The value that the operator operates on is called the operand.

Operators are divided into:

- Arithmetic Operators
- Relational Operators
- Assignment Operators
- Logical Operators
- Bitwise Operators
- Membership Operators
- Identity Operators

We are going to learn some of them. Rest are your homework :)

```python
## Arithmetic Operators

print(2 + 3) 
print(1 - 2)
print (5 * 5)
print (7 / 3)
print (11 % 4)
```

```python
## Relational Operators

# Equality
print(1 == 0)

# Greater Than Operator
print(4 > 3)

# Not equal to operator
print( 4 != 4)
```

```python
# Assignment Operators

a=1
print(a)

b=0
b+=1 # Shorthand operator
print(b)
```

```python
b=0
b=b+1
print(b)

b+=
```

```python
# Logical Operators

a=1
b=0

print(a and b)
print(a or b)
print(not a)
```

```python
# Membership Operators

series = [10,20,30]
print(10 in series)
print(10 not in series)
```

```python
# Identity Operators

a = 40
b = 20
c = 40

print(a is b)
print(a is not b)
print(a is c)
```

**The above are some basic operators that we will use in Python. As we learn more, we will learn more complex operators.**
